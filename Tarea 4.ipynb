```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
exo=pd.read_csv("exoplanets.csv")           #Se lee el archivo y se muestran los primeros datos
exo.head(8)
```

```python
#Acá separo las columnas que se piden

exoN=pd.DataFrame()

NAME=exo['NAME']
TEFF=exo['TEFF']
MASS=exo['MASS']
A=exo['A']
DENSITY=exo['DENSITY']
R=exo['R']
STAR=exo['STAR']
MSTAR=exo['MSTAR']
RSTAR=exo['RSTAR']
BINARY=exo['BINARY']


exoN["NAME"]=NAME
exoN['TEFF']=TEFF
exoN['MASS']=MASS
exoN['A']=A
exoN['DENSITY']=DENSITY
exoN['R']=R
exoN['STAR']=STAR
exoN['MSTAR']=MSTAR
exoN['RSTAR']=RSTAR
exoN['BINARY']=BINARY

exoN.head(11)
```

```python
#Las pongo en el orden que están en el documento y quito los sistemas binarios

EXO=pd.DataFrame()

for i in range(0,5747):
    A=exoN.iloc[i]
    if A['BINARY']==0:
        EXO=EXO.append(A)
        

columns_titles = ["NAME","TEFF",'MASS','A','DENSITY','R','STAR','MSTAR','RSTAR','BINARY']
Exop=EXO.reindex(columns=columns_titles)
```

```python
#Acá se calcula el radio de los planetas en radios de la tierra

qM = (1.898E27)/(5.972E24)   #Factor multiplicativo para pasar de masa en masas de Jupiter a masas de la Tierra
MASSE=Exop['MASS']
MASSE=MASSE*qM
Exop['MASSE']=MASSE          #Agregamos esta nueva columna al Data Frame


qR=(69911)/(6371)            #Factor multiplicativo para pasar de  en radios de Jupiter a radios de la Tierra
RE=Exop['R']
Exop['RE']=RE*qR             #Agregamos esta nueva columna al Data Frame

#Exop.head(5)
```

```python
#Acá se calcula la luminosidad
#Debemos tener en cuenta que MSTAR y RSTAR están expresados en proporción a masa y radio del sol


RSTAR=Exop['RSTAR']                         #Separamos las columnas para calular la luminosidad
T=Exop['TEFF']
Rs=6.96E8  
Rt=6.378e6                                  #Definimos el radio del sol en metros
Sigma=(5.68E-8)                             #Definimos la constante de Steffan-Boltzmann en W m^-2 K^-4
Rm=RSTAR*(Rs)                               #Dejamos la columna de radios con unidades en metros
#print(Rm**2)
L=4*np.pi*(Rm**2)*Sigma*(T**4)              #Calculamos la columna de luminosidad
Exop['LUM']=L

Exop.head(5)
```
